# Implementation of the Zel'dovich approximation
In this notebook, you will use the Zel'dovich approximation to evolve a particle distribution over time.

We will:
1. Set up an **initial grid of particles**.
2. Define an **initial perturbation** using a sine wave.
3. Compute **displacements** using the Zel'dovich Approximation.
4. Visualize the **evolution of density**.



### Step 1: Set Up the Initial Conditions
We assume a **1D universe** with particles initially at rest.

````python
# Define grid size and number of particles
L = 10.0  # Size of the 1D domain
N = 100    # Number of particles
x_initial = np.linspace(0, L, N)  # Initial positions
````


### Step 2: Define an Initial Perturbation
Define the displacement field according to the exercise sheet


````python
### Step 2: Define an Initial Perturbation
Psi = np.sin(x_initial) # Displacement field
````

### Step 3: Apply the Zel'dovich Approximation
Use the Zel'dovich approximation to calculate ```x``` as a function of ```x_initial```, ```Psi```and time ```t```

````python
### Step 3: Apply the Zel'dovich Approximation
def x(x_initial, Psi, t):
    x = x_initial + t * Psi
    return x
````

### Step 4: Visualize particle distribution over time

````python
# Time evolution of displacement 
times = np.linspace(0, 10, 20) 

x_over_time=[]
for t in times:
    x_over_time.append(x(x_initial, Psi, t))
````

````python
plt.figure(figsize=(12,12))
# Plotting
colors = plt.cm.jet(np.linspace(0,1,len(x_initial)))
for i, x in enumerate(x_over_time):
    plt.scatter(x, np.ones_like(x) * times[i], label=f't={times[i]}', c='r', s=0.5)
    # plt.scatter(x[-1], times[i], label=f't={times[i]}')


# plt.legend()
plt.xlabel("Position x")
plt.ylabel("Time t")

````
